```python
import seaborn as sns
iris= sns.load_dataset('iris')
iris.head()
```

```python
%matplotlib inline
sns.set()
sns.pairplot(iris, hue='species', size=1.5)
```

```python
#Drop the Species column for the train data
X_iris = iris.drop('species', axis=1)
X_iris.shape
```

```python
#We only need this column for the prediction
y_iris = iris['species']
y_iris.shape
```

```python
import matplotlib.pyplot as plt
import numpy as np

#Random generator with the choosen state
rng=np.random.RandomState(42)

#rand gives random values in a given shape 
x=10*rng.rand(50)
y=2*x - 1 + rng.randn(50)

plt.scatter(x,y)
```

```python
from sklearn.linear_model import  LinearRegression
```

```python
model = LinearRegression(fit_intercept=True)
model
#let it calculate the intercept, if false, we should have a centerized data
```

```python
X=x[:, np.newaxis]
X.shape
```

```python
x #Series
```

```python
X #50x1 matrix
```

```python
model.fit(X,y)
```

```python
model.coef_ #Slope
```

```python
model.intercept_ #Intercept
```

```python
xfit=np.linspace(-1,11)
xfit
#new set of data
```

```python
#transform X from series to a matrix
Xfit=xfit[:,np.newaxis]
#predict the y values for the new X values
yfit=model.predict(Xfit)
```

```python
#Plot the data
plt.scatter(x,y)
plt.plot(Xfit,yfit);
```

**Supervised Learning example** 

```python
from sklearn.cross_validation import train_test_split
Xtrain, Xtest, ytrain, ytest = train_test_split(X_iris, y_iris, random_state=1)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
model.fit(Xtrain, ytrain) #fit the model
y_model = model.predict(Xtest) #predict on new data
```

```python
#For evaluation we import the accuracy_score from metrics
from sklearn.metrics import accuracy_score
accuracy_score(ytest,y_model)
```

**Unsupervised learning example**

```python
from sklearn.decomposition import PCA
model = PCA(n_components= 2)
model.fit(X_iris)

#Transform the dataset to a 150*2 matrix
X_2D= model.transform(X_iris)
```

```python
X_2D.shape
```

```python
iris['PCA1'] = X_2D[:,0]
iris['PCA2'] = X_2D[:,1]
sns.lmplot("PCA1","PCA2", hue='species', data=iris, fit_reg=False);
#Plot the 2 dimensional representation of the dataset
```

**Unsupervised learning: Iris clustering**

```python
from sklearn.mixture import GMM
model= GMM(n_components=3, covariance_type='full')
model.fit(X_iris)
y_gmm = model.predict(X_iris)
```

```python
iris['cluster'] = y_gmm
sns.lmplot('PCA1','PCA2', data = iris, hue='species', col='cluster', fit_reg=False)
```

**Exploring Handwritten Digits**

```python
from sklearn.datasets import load_digits
digits = load_digits()
digits.images.shape
```

```python
fig, axes = plt.subplots(10,10,figsize=(8,8),subplot_kw={'xticks':[], 'yticks':[]}
                        ,gridspec_kw=dict(hspace=0.1,wspace=0.1))

for i, ax in enumerate(axes.flat): #.flat changes the aray to 1D
    ax.imshow(digits.images[i], cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(digits.target[i]),
           transform=ax.transAxes, color='green')
    
```

```python
X = digits.data
X.shape
```

```python
y = digits.target
y.shape
```

```python
from sklearn.manifold import Isomap
iso = Isomap(n_components=2) #reduce the dimension to 2
iso.fit(digits.data)
data_projected = iso.transform(digits.data)
data_projected.shape
```

```python
plt.scatter(data_projected[:,0],data_projected[:,1],
           c=digits.target, edgecolor='none',alpha=0.5, cmap=plt.cm.get_cmap('spectral', 10))
plt.colorbar(label='digit label', ticks=range(10))

#sets the color limit
plt.clim(-0.5,9.5)
            
```

**Classification on digits**

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,random_state=0)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
model.fit(Xtrain,ytrain)
y_model = model.predict(Xtest)
```

```python
accuracy_score(ytest,y_model)
```

```python
from sklearn.metrics import confusion_matrix

mat= confusion_matrix(ytest,y_model)

#with the use of heatmap it's easy to visualize where our algorithm was weak
sns.heatmap(mat, square=True, annot= True, cbar=False)
plt.xlabel('predicted value')
plt.ylabel('true value')
```

```python
fig, axes=plt.subplots(10,10,figsize=(8,8),
                      subplot_kw={'xticks':[], 'yticks':[]}
                      ,gridspec_kw=dict(hspace=0.1,wspace=0.1))

test_images = Xtest.reshape(-1,8,8)

for i, ax in enumerate(axes.flat):
    ax.imshow(test_images[i],cmap='binary', interpolation='nearest')
    ax.text(0.05, 0.05, str(y_model[i]),
                           transform = ax.transAxes,
                           color='green' if (ytest[i]==y_model[i]) else 'red')
```
